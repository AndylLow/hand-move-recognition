```python
# TensorFlow and tf.keras
import tensorflow as tf
import os
# Helper libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
from sklearn.model_selection import train_test_split

print(tf.__version__)
```

```python
def reduce_to_single_record(group):
    # Create a new dataframe with a single row and 2 columns
    reduced_df = pd.DataFrame(columns=['başlık1', 'başlık2'])
    
    # Add the data from the group to the new dataframe
    reduced_df['başlık1'] = group['başlık1'].values
    reduced_df['başlık2'] = group['başlık2'].values
    #reduced_df['title']=group['title'].unique()[0]
    
    return reduced_df
```

```python
filenames = glob.glob('/content/*.csv')
#dfs = [pd.read_csv(f, sep=',', names=['başlık1', 'başlık2']) for f in filenames]
dfs= []
for f in filenames:
  dfr=pd.read_csv(f, sep=',', names=['başlık1', 'başlık2'])
  dfr = dfr.groupby(dfr.index // 4000).apply(lambda x: reduce_to_single_record(x)) #YENİ SATIR
  dfr['title'] = f.replace("/content/","")[:3]
  dfs.append(dfr)
df = pd.concat(dfs)
print(df['title'])
```
